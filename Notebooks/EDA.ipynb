```python
#so,change the names of the column. Because there may be problems for future analysis.
df=df.rename(columns={'country':'Country',
                        'year':'Year',
                        'sex':'Gender',
                        'age':'Age',
                        'suicides_no':'SuicidesNo',
                        'population':'Population',
                        'suicides/100k pop':'Suicides100kPop',
                        'country-year':'CountryYear',
                        'HDI for year':'HDIForYear',
                        ' gdp_for_year ($) ':'GdpForYearMoney',
                        'gdp_per_capita ($)':'GdpPerCapitalMoney',
                        'generation':'Generation'
                        })
```

```python
df.columns
```

```python
#how many rows and columns are there for all data?
print(f"This data set has {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
#check any column has null values??
df.isnull().values.any()
```

```python
#print the column name which has missing values and number of missing values 
df.isnull().sum()[lambda x: x>0]
```

```python
#As you can see, most of the HDIForYear value is empty. That's why I want this value deleted.
df=df.drop(['HDIForYear','CountryYear'],axis=1)
```

```python
#Now start analysis, min year and max year will find them
min_year=min(df['Year'])
max_year=max(df['Year'])
print('Min Year :',min_year)
print('Max Year :',max_year)
```

```python

#1985 min year,2016 max year.

df_1985 = df[(df['Year'] == min_year)]
   
ax = sns.catplot(data = df_1985,
                    y = 'Country',
                    hue = 'Gender',
                    kind = 'count',
                    hue_order = ['male','female'],
                    height = 25,
                    legend  =  False)
ax.set_xticklabels(fontsize = 20)
ax.set_yticklabels(fontsize = 20)
plt.ylabel('Countries',fontsize = 25)
plt.xlabel('Count Male vs Female',fontsize = 25)
plt.title('1985 Year Gender wise Suicide Rate',fontsize = 35)
plt.legend(title ="Gender",fontsize = '20', title_fontsize = '30')
plt.show()
```

```python
#Very odd all the rates came on an equal level.
df_2016 = df[(df['Year'] == max_year)]
   
ax = sns.catplot(data = df_2016,
                    y = 'Country',
                    hue = 'Gender',
                    kind = 'count',
                    hue_order = ['male','female'],
                    height = 20,
                    legend  =  False)      
ax.set_xticklabels(fontsize = 20)
ax.set_yticklabels(fontsize = 20)
plt.title('2016 Year Gender wise Suicide Rate',fontsize = 35)
plt.ylabel('Countries',fontsize = 25)
plt.xlabel('Count Male vs Female',fontsize = 25)
plt.legend(title ="Gender",fontsize = '20', title_fontsize = '30')
plt.show()
```

```python
# plt.figure(figsize=(9,6))
ax = sns.lineplot(data=df,
                    x = 'Year',
                    y = 'SuicidesNo',
                    hue ='Gender',
                    hue_order = ['male','female'],
                    ci = None,
                    linewidth = 2,
                    legend = False)  
plt.title('Suicide number by Gender (1985-2016)', fontsize=15)
plt.legend(["male", 'female'])
plt.show()
```

```python
plt.figure(figsize=(9,6))
ax = sns.lineplot(data=df,
                    x = 'Year',
                    y = 'SuicidesNo',
                    hue ='Age',
                    hue_order = ['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years'],
                    ci = None,
                    linewidth = 2,
                    style = "Age",
                    dashes=False) 
plt.title('Suicide number by Age (1985-2016)', fontsize=20)
plt.show()
```
